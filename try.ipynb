```python
from library.ron_utils import *
from torch_lib.KITTI_Dataset import FrameCalibrationData
for i in range(10):
    lines = open('Kitti/training/label_2/%06d.txt'%i).readlines()
    cam_to_img = FrameCalibrationData('Kitti/training/calib/%06d.txt'%i)
    for j in range(len(lines)):
        elements = lines[j].strip().split()
        if elements[0] == 'DontCare':
            continue
        for i in range(1, len(elements)):
            elements[i] = float(elements[i])
        trun = elements[1]
        alpha = elements[3]
        box2d = [[int(elements[4]),int(elements[5])], [int(elements[6]),int(elements[7])]]
        dimension = [elements[8], elements[9], elements[10]]
        location = [elements[11], elements[12], elements[13]]
        location[1] -= dimension[2]/2
        theta_ray = elements[14] - elements[3]
        Ry = elements[14]
        proj_matrix = cam_to_img.p2
        obj_W, obj_L = dimension[1], dimension[2]
        calc_depth = round(calc_depth_with_alpha_theta(1242, box2d, proj_matrix, obj_W, obj_L, alpha, 0), 2)
        print(elements[0], location[2], calc_depth, abs(location[2]-calc_depth))
        visual_W = abs(obj_L*np.cos(alpha)) + abs(obj_W*np.sin(alpha))
        Wview = (visual_W)*(1242/get_box_size(box2d)[0])
        print(Wview, obj_W)
```
