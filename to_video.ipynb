```python
import cv2
import os
def image_to_video():
    file = './origin_dim/'  # 图片目录
    output = 'no_fix.avi'  # 生成视频路径
    height = 740
    weight = 1226
    fps = 5
    # fourcc = cv.VideoWriter_fourcc('M', 'J', 'P', 'G') 用于avi格式的生成
    fourcc = cv2.VideoWriter_fourcc(*'XVID')  
    videowriter = cv2.VideoWriter(output, fourcc, fps, (weight, height))  # 创建一个写入视频对象
    for file_name  in sorted(os.listdir(file)):
        path = file + file_name
        print(path)
        frame = cv2.imread(path)
        videowriter.write(frame)

    videowriter.release()

image_to_video()

```
